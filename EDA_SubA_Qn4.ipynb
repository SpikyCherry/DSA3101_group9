```python
# Load Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/alvislow/repos/DSA3101/bank_customers_train.csv")
```

```python
print(df.head())
```

```python
# Remove duplicates
print(df.duplicated().sum())
df = df.drop_duplicates() 
```

```python
# Boxplot for outliers
plt.figure(figsize=(10, 5))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x=df['marital'], y=df['y'].apply(lambda x: 1 if x == 'yes' else 0))
plt.title("Subscription Rate by Marital Status")
plt.show()
```

```python
sns.barplot(x=df['month'], y=df['y'].apply(lambda x: 1 if x == 'yes' else 0))
plt.xticks(rotation=90)
plt.title("Subscription Rate by Month")
plt.show()

sns.barplot(x=df['day_of_week'], y=df['y'].apply(lambda x: 1 if x == 'yes' else 0))
plt.title("Subscription Rate by Day of the Week")
plt.show()
```

```python
sns.boxplot(x=df['job'], y=df['age'], hue=df['y'])
plt.xticks(rotation=90)
plt.title("Age Distribution by Job & Subscription Status")
plt.show()
```
