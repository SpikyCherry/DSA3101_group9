## Pie Charts for categorical data

```python
gender_counts = data['Gender'].value_counts()
education_counts = data['Education_Level'].value_counts()
married_counts = data['Marital_Status'].value_counts()
```

```python
colors = ['lightblue','lightcoral']
explode = [0.1, 0]
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)
```

```python
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
```

```python
plt.pie(married_counts, labels= married_counts.index, autopct='%1.1f%%', startangle=140)
```

```python
attrition_counts = data['Attrition_Flag'].value_counts()
plt.pie(attrition_counts, labels= attrition_counts.index, autopct='%1.1f%%', startangle=140)
```

```python
income_counts = data['Income_Category'].value_counts()
plt.pie(income_counts, labels= income_counts.index, autopct='%1.1f%%', startangle=140)
```

```python
card_counts = data['Card_Category'].value_counts()
plt.pie(card_counts, labels= card_counts.index, autopct='%1.1f%%', startangle=140)
```
